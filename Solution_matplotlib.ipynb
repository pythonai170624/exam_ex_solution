```python
import matplotlib.pyplot as plt
import pandas as pd

################### 1
# Data
data = {
    'Manufacturer': ['Toyota', 'Ford', 'Ford', 'Audi', 'BMW', 'Toyota', 'Audi', 'BMW', 'Nissan', 'Nissan'],
    'Type': ['Sedan', 'Truck', 'SUV', 'Sedan', 'Sedan', 'Truck', 'SUV', 'Truck', 'SUV', 'Sedan'],
    'Fuel_Efficiency': [34, 22, 24, 27, 28, 24, 26, 21, 23, 35],
    'Horsepower': [180, 320, 260, 220, 240, 290, 210, 310, 250, 200],
    'Price': [30000, 40000, 35000, 45000, 50000, 32000, 47000, 48000, 33000, 31000]
}

# Convert to DataFrame
df = pd.DataFrame(data)

#################### 2
# Plot a line chart representing the 'Horsepower' vs 'Price' relationship for all
# manufacturers.
# Use title, axis labels and axis limitation. Make your line chart dashed style (--),
# in red color and add the actual data points on the chart itself (marker).

#fig = plt.figure(figsize=(8,4)) 
custom_colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'pink', 'gray', 'olive', 'cyan']
fig, ax = plt.subplots(len(df['Manufacturer'].unique()), figsize=(8, 20))

df = df.sort_values(by='Horsepower')

for i, manufacturer in enumerate(df['Manufacturer'].unique()):
    manufacturer_data = df[df['Manufacturer'] == manufacturer]
    ax[i].plot(manufacturer_data['Horsepower'], manufacturer_data['Price'], '--', marker='o', label=manufacturer, color=custom_colors[i])
    ax[i].set_title(f"Manufacturer: {manufacturer}")
    ax[i].set_xlabel('Horsepower')
    ax[i].set_ylabel('Price')
    ax[i].legend()

# for manu in df['Manufacturer'].values:
#     plt.plot(df[df['Manufacturer'] == manu]['Horsepower'], df[df['Manufacturer'] == manu]['Price'])
plt.tight_layout(h_pad=0.6)  # change to 2
plt.show()
```
