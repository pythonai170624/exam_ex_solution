```python
# Exercise 1:
# 1 
# create the df

# 2
# Create a new column could ‘classification’ and classify the customers based
# on their `TotalPurchaseAmount`. If the purchase amount is less than 500 it's
# 'Low', otherwise it's 'High'.
# 3. Using the column with the previous exercise create a new column could
# ‘‘classification_by_age’ and put values according to the following logic:
# If the classification ‘Low’ and the customer age is below 20 put ‘Low_Young’
# If the classification ‘HIGH’’ and the customer age is above 20 put ‘HIGH_Old’
# If the classification ‘Low’ and the customer age is above 20 put ‘Low_Old’
# If the classification ‘High’ and the customer age is below 20 put ‘High_Young’

df = pd.DataFrame({
    'CustomerID': np.arange(1, 11),
    'FirstName': ['John', 'Steve', 'Anna', 'Mike', 'Emily', 'Jake', 'Laura', 'Nick', 'Olivia', 'David'],
    'LastName': ['Smith', 'Johnson', 'Williams', 'Jones', 'Brown', 'Davis', 'Miller', 'Wilson', 'Moore', 'Taylor'],
    'Age': np.random.randint(25, 65, size=10),
    'TotalPurchaseAmount': np.random.randint(100, 1000, size=10)
})

df['classification'] = df['TotalPurchaseAmount'].apply(lambda total: 'Low' if total < 500 else 'High')

def age_desc(age, classification):
    return 'Low_' + classification if age < 20 else 'High_' + classification

df['classification_by_age'] = np.vectorize(age_desc)(df['Age'], df['classification'])

df
```
