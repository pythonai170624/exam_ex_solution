```python
# 3
# Create a dataframe with 4 rows and 6 columns populated with random
# numbers between 0,100 - use seed 100 for the exercise.
# 2. Give different name for each column in the dataframe
# 3. Add a new column with the sum of each row
# 4. Calculate the sum of each column in the dataframe
# 5. Find the maximum and minimum value of each row and each column.
# 6. Create a copy of the original dataframe and replace all even numbers in the
# copy dataframe with the value -1

df = pd.DataFrame(np.random.randint(0, 101, (4, 6)), columns= ['zero', 'one', 'two', 'three', 'four', 'five'])
#display(df)
df['sum_rows'] = df.apply(lambda row: row.sum(), axis=1)  # row is Series of Index-value 
    # i.e.
    # index        value
    # zero         54
    # one          26
    # two          66
    # three        57
    # four         99
    # five         47
    # if we sum it will sum all of the values, could also use np.sum instead of lambda row: row.sum()
display(df)
df = df.drop(columns='sum_rows')

sum_column = df.sum(axis=0)
display('sum', sum_column)

max_in_row = df.max(axis=1)
display('max in row', max_in_row)

max_in_column = df.max(axis=0)
display('max in col', max_in_column)

df_even = df.copy()
df_even[df % 2 == 0] = -1
display(df_even)
```

```python
# 4
# Create three Pandas DataFrames, each representing the car inventory of
# three different dealerships (A, B, C). Each dealership should have at least five
# different car models on hand, along with the corresponding amount in stock.
# For example - Dealership A → Has 5 units of Ford Mustang GT in-stock
# 2. Write a Python function that gets a list of cars to buy and for each car check
# where the available dealerships that are selling this car and have it in stock.
# Your function should return for each car a list of the available dealerships and
# decrease the amount in stock for the relevant dealerships.

# * In case there are more than 1 available dealership that sell the requested
# car decrease the stock amount for only one of the available dealerships

dealership_a = pd.DataFrame({
    'Car Model': ['Ford Mustang GT', 'Chevrolet Camaro', 'Toyota Corolla', 'Honda Accord', 'Chevrolet Corvette'],
    'Stock': [5, 3, 10, 7, 4]
})

dealership_b = pd.DataFrame({
    'Car Model': ['Ford Bronco', 'Jeep Cherokee', 'Subaru Outback', 'Tesla Model 3', 'Nissan Altima'],
    'Stock': [2, 4, 8, 5, 6]
})

dealership_c = pd.DataFrame({
    'Car Model': ['Hyundai Sonata', 'Kia Soul', 'Jeep Wrangler', 'Chevrolet Tahoe', 'Subaru Outback'],
    'Stock': [6, 7, 3, 4, 5]
})

dealerships = [dealership_a, dealership_b, dealership_c]

display(dealership_a, dealership_b, dealership_c)
def buy_car(cars_list):
    result = dict()
    for car in cars_list:
        available_dealerships = []
        already_decrease = False
        for dealership in dealerships:
            #   'Kia Soul' in dealership_c['Car Model'].values
            # or
            #   any(dealership_c['Car Model'].isin(['Kia Soul']))
            if car in dealership['Car Model'].values:
                if not already_decrease:
                    dealership.loc[dealership['Car Model'] == car, 'Stock'] -= 1
                    already_decrease = True
                available_dealerships.append(dealership)
        result[car] = available_dealerships
    return result

result = buy_car(['Kia Soul', 'Subaru Outback'])
for key, value in result.items():
    print(key, ': \n')
    for v in value:
        print(v)
    print()
print('Done')

#dealership_c['Car Model']
```
